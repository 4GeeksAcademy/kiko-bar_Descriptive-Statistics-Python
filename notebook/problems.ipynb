## Exercise 1 – How long are movie titles?

In this exercise, you will apply descriptive statistics techniques to a real dataset: a database of 1000 popular movies from the IMDb website.

From the title of each movie, you will:
- Calculate the title length (in characters).
- Obtain key statistical metrics: mean, median, mode, range, variance, standard deviation, skewness, and kurtosis.
- Identify the longest and shortest titles.
- Visualize the distribution of these lengths with a histogram and a density curve.


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Load IMDb dataset
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Create a new column called 'title_length' with the length (number of characters) of each title
df['title_length'] = df['title'].str.len()
df['title_length'].describe()
```

```python
df['title_length']
```

```python
# Calculate mean, median, and mode of the 'title_length' column
mean = df['title_length'].mean()
median = df['title_length'].median()
mode = df['title_length'].mode()
print(f"The mean: {mean}\nThe median: {median}\nThe mode: {mode}")
```

```python
# Calculate the range, variance, and standard deviation of 'title_length'
range = df['title_length'].max() - df['title_length'].min()
variance = df['title_length'].var()
standard = df['title_length'].std()
print(f"Variance: {variance}\nStandard Deviation: {standard}\nRange: {range}")

```

```python
# Calculate skewness and kurtosis of the 'title_length' column
skewness = df['title_length'].skew()
kurt = df['title_length'].kurt()
print(f"Kurtosis: {kurt}\nSkewness: {skewness}")
```

```python
# Print the shortest and longest title based on their length
df['title_length'].idxmax()
df['title_length'].idxmin()
print(f"The loooongest title: {df['title'].iloc[df['title_length'].idxmax()]}\nThe shortest title: {df['title'].iloc[df['title_length'].idxmin()]}")
```

```python
# plot your results
df['title_length'].hist(bins=32)
plt.title('Histogram of the Value Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

# Conclusion for excercise 1:
- We can see most of the data is concentrated on the left side with a long tail that spread toward the higher values
- The highest frequency occurs in the bin located between 10 - 15. Close where the mode is.
- There is a possitive relation ship between the mode, median and mean value which follow this pattern mode < median < mean
- The hist suggest that the tipical long character for a title is around 14 which is the tendency but not necessaryly will always follow this as we can see longer title in time.

## Exercise 2 – Manual Calculation of Standard Deviation

In this exercise, you will manually calculate the standard deviation by selecting five actual titles from the movie dataset you loaded earlier.


```python
# Create a list with 5 actual values from df['title_length'], for example: [10, 13, 14, 18, 22]
my_lst = [13, 22, 15, 12, 11]
```

```python
# Calculate the mean of the 5 values
mean = sum(my_lst) / len(my_lst)
print(f"Calculate the Mean: {mean}")
```

```python
# Calculate the squared differences with respect to the mean
difference = [x - mean for x in my_lst]
print(f"Calculate the Differences: {difference}")
```

```python
square_difference = [x ** 2 for x in difference]
print(f"List of square values: {square_difference}")
```

```python
# Calculate the variance (average of the squared differences) and its square root to obtain the standard deviation
square_difference_mean = sum(square_difference) / (len(square_difference) - 1)
square_root = (square_difference_mean) ** 0.5
print(f"This is the result of the manual calculation for Standard Deviation: {square_root}")
```
